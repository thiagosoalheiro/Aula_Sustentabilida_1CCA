```python
# IMPORTANDO BIBLIOTECA PARA MANIPULAÇÃO DE TABELAS
import pandas as pd
```

```python
tabela = pd.read_csv('/content/BASE_DE_DADOS.csv')
```

```python
# VISUALIZANDO PRIMEIROS REGISTROS DA TABELA
  # LINHAS ---> " registros "
  # COLUNAS ---> " atributos " ---> " features "
tabela.head()
```

```python
tabela.columns
```

```python
# ESTATISTICAS DESCRITIVAS DAS COLUNAS
# EXEMPLO: VOLTAGE (TENSÃO, EM VOLTS) E GLOBAL_INTENSITY (CORRENTE, EM AMPÉRES)

# VOLTAGE

# MÉDIA ARITMÉTICA
media_tensao = tabela['Voltage'].mean()
print(f"A média da tensão será de: {media_tensao:.2f}V")

# MEDIANA
mediana_tensao = tabela['Voltage'].median()
print(f"A mediana da tensão será de: {mediana_tensao:.2f}V")

# DESVIO PADRÃO
desvio_padrao = tabela['Voltage'].std()
print(f"O desvio padrão da tensão será de: {desvio_padrao:.2f}V")
```

```python
# GLOBAL_INTENSITY

# MÉDIA ARITMÉTICA
media_corrente = tabela['Global_intensity'].mean()
print(f"A média da corrente será de: {media_corrente:.2f}A")

# MEDIANA
mediana_corrente = tabela['Global_intensity'].median()
print(f"A mediana da corrente será de: {mediana_corrente:.2f}A")

# DESVIO PADRÃO
desvio_corrente = tabela['Global_intensity'].std()
print(f"O desvio padrão da corrente será de: {desvio_corrente:.2f}A")
```
